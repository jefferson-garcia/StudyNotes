**Observación:** Del teorema $1.1.4$ podemos saber cuándo una sucesión de numeros determinan un grafo no simple; sin embago, queda la pregunta: ¿cuáles son las condiciones necesarias para determinar un grafo simple?

La pregunta anterior lleva a lo siguiente:


Dada una sucesión $(d_1, d_2, \cdots, d_n)$ de enteros positivos, y no creciente, ¿cómo hacemos para determinar si los valores de esta sucesión determinan un grafo simple?

La respuesta a este problema la dan Havel(1955) y Hakimi(1961) con la implementación de un algoritmo recursivo que tiene la siguiente estructura:


**Input:** Sucesión $(d_1, d_2, \cdots, d_n)$ decreciente de números positivos.

**Output:** True o False indicando si la sucesión representa o no un grafo simple.


El pseudocódigo y teoría necesaria para llegar a la implementación del algoritmo se encuentran en las páginas $9$ y $10$ del libro.

```python
# Implementación del algoritmo de Havel-Hakimi.

def havelHakimi(secuencia):
    secuencia.sort(reverse=True)
    if secuencia[0] >= len(secuencia) or secuencia[-1] < 0:
        return False
    elif secuencia[0] == 0:
        return True
    else:
        valor_quitado = secuencia.pop(0)
        for i in range(valor_quitado):
            secuencia[i] -= 1
        secuencia.sort(reverse=True)

    return havelHakimi(secuencia)
```
